# CREATING FILTERS AND WORKING WITH THEM

```python
filt = (dict_df['last_name'] == 'Gupta')

dict_df.loc[filt]
```

```python
dict_df.loc[~filt]
```

```python
names = ['Gupta','Narula']

country_filter = (dict_df['last_name'].isin(names))

dict_df.loc[country_filter]
```
