# deleting rows and Columns

```python
import pandas as pd
df = pd.read_csv('/content/survey_results_public.csv')
schema_df = pd.read_csv('/content/survey_results_schema.csv')
```

```python
people = {
    'first' : ['Corey','Jane','John'],
    'last' : ['Schafer','Doe','Doe'],
    'email' : ['CoreySchafer@gmail.com','JaneDoe@gmail.com','JohnDoe@gmail.com'],
}

people_df = pd.DataFrame(people)
```

```python
people_df
```

```python
people_df['first'] + ' ' + people_df['last']
```

```python
people_df
```

```python
#just like a dictionary
#if the key is not present it creates a new set of key value
people_df['full_name'] = people_df['first'] + ' ' + people_df['last']
```

```python
people_df
```

```python
#removing coluumns in our dataFRAME
# people_df.drop(columns = ['first','last'])
people_df.drop(columns = ['first','last'], inplace = True)
```

```python
people_df
```

```python
#reversing - split the full name into first and last
people_df['full_name'].str.split(' ',expand = True)
```

```python
people_df[['first','last']] = people_df['full_name'].str.split(' ',expand = True)
```

```python
people_df
```
