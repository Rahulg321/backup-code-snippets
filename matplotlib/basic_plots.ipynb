```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('Our First Graph')
plt.show()
```

```python
#resizing our titles and labels

x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]

plt.plot(x,y)
plt.xlabel('X AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.ylabel('Y AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.title('Our First Graph', fontdict = {'fontname':'Comic Sans MS','fontsize':20})
plt.show() 
```

```python
# Removing integers on our x axis

x = [0,1,2,3,4,]
y = [0,2,4,6,8,]

plt.plot(x,y)
plt.xlabel('X AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.ylabel('Y AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.title('Our First Graph', fontdict = {'fontname':'Comic Sans MS','fontsize':20})

plt.xticks([0,1,2,3,4,5])  # tics are the axis values/points
plt.yticks([0,2,4,6,8,10])
plt.show() 
```

```python

x = [0,1,2,3,4,]
y = [0,2,4,6,8,]

plt.plot(x,y,label = '2x', color = 'red', linewidth = 2, marker = '*', markersize = 10)  #label is our slope of the graph
plt.xlabel('X AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.ylabel('Y AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.title('Our First Graph', fontdict = {'fontname':'Comic Sans MS','fontsize':20})

plt.xticks([0,1,2,3,4,5])  
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show() 
```

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,label = '2x', color = 'black', linewidth = 4, marker = '.', markersize = 14)
plt.plot(x,y,label = '2x', color = 'pink', linewidth = 5, marker = '*', markersize = 15)
plt.title('My Second Graph', fontdict={'fontname':'Comic Sans Ms','fontsize':18})
plt.xlabel('X Axis', fontdict={'fontname':'Comic Sans Ms','fontsize':16})
plt.ylabel('Y Axis', fontdict={'fontname':'Comic Sans Ms','fontsize':16})
plt.xticks([1,2,3,4,5])
# plt.yticks([2,4,6,8,10])
plt.legend()
plt.show()
```

```python

x = [0,1,2,3,4,]
y = [0,2,4,6,8,]


#shorthand Notation
#[[color],[marker],[line]]
plt.plot(x,y, 'r*-', label = '2x',color = 'pink')  #label is our slope of the graph
plt.xlabel('X AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.ylabel('Y AXIS',fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.title('Our First Graph', fontdict = {'fontname':'Comic Sans MS','fontsize':20})

plt.xticks([0,1,2,3,4,5])  
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show() 
```
