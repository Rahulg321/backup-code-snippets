```python
from matplotlib import pyplot as plt
```

```python
# plt.style.use('fivethirtyeight')

plt.title('My Awesome Pie Chart', fontdict={'fontname':'Comic Sans Ms','fontsize':20})

slices = [60,40]

plt.pie(slices)
plt.tight_layout()
plt.show()
```

```python
#adding labels for our slices 

slices = [60,40]

plt.pie(slices,labels = ['sixty','forty'])

plt.title('My Awesome Pie Chart', fontdict={'fontname':'Comic Sans Ms','fontsize':20})
 
plt.tight_layout()
plt.show()
```

```python

slices = [60,40]

plt.pie(slices,labels = ['sixty','forty'], wedgeprops={'edgecolor':'black'})

plt.title('My Awesome Pie Chart', fontdict={'fontname':'Comic Sans Ms','fontsize':20})
 
plt.tight_layout()
plt.show()
```

```python
slices = [60,40,30,20]
colors = ['blue','red','yellow','green']
labels = ['sixty','forty', 'Extra 1', 'Extra 2']
plt.pie(slices,labels = labels,colors = colors, wedgeprops={'edgecolor':'black'})

plt.title('My Awesome Pie Chart', fontdict={'fontname':'Comic Sans Ms','fontsize':20})
 
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', ]

plt.pie(slices,labels = labels, wedgeprops={'edgecolor':'black'})

plt.title('My Awesome Pie Chart', fontdict={'fontname':'Comic Sans Ms','fontsize':20})
 
plt.tight_layout()
plt.show()
```

```python
#emphaisis on a single chart

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', ]

explode_list = [0,0,0,0.2,0] #list of floats that will say how much we want to emphasise that slice

plt.pie(slices,labels = labels, explode = explode_list,shadow = True,autopct ='%1.1f%%', wedgeprops={'edgecolor':'black'},startangle=40)

plt.title('My Awesome Pie Chart', fontdict={'fontname':'Comic Sans Ms','fontsize':20})
 
plt.tight_layout()
plt.show()
```
