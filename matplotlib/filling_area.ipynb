```python
import pandas as pd
from matplotlib import pyplot as plt
```

we will learn to fill learn to fill data in our line plots, whichever way we want to 

we can fill beneath a certain line plot or above a line plot 

```python
data = pd.read_csv('C:/Users/Lenovo\OneDrive/Desktop/LIBRARIES/MATPLOTLIB/WORKING_MODULES/line_data.csv')
```

```python
data.head()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
#filling a data beneath a python line plot  

plt.plot(ages,dev_salaries,color = '#444444',linestyle ='--',label = 'Alldevs')
plt.plot(ages,py_salaries,label = 'Python')


#alpha arguement -> how much we can see through our fill 
plt.fill_between(ages,py_salaries, alpha = 0.25)  #fills between python line plot and all the way down to the xline 


plt.legend()
plt.title('Median Salary(USD) by Ages', fontdict={'fontname':'Comic Sans Ms', 'fontsize':20})
plt.xlabel('Ages',fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})
plt.ylabel('Median Salary', fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})

plt.tight_layout()
plt.show()
```

```python
#filling a data beneath a python line plot  

plt.plot(ages,dev_salaries,color = '#444444',linestyle ='--',label = 'Alldevs')
plt.plot(ages,py_salaries,label = 'Python')

overall_median = 52787

#the third arguement speicifies where exactly we want our fill to cut through
plt.fill_between(ages,py_salaries, overall_median, alpha = 0.25)   

plt.legend()
plt.title('Median Salary(USD) by Ages', fontdict={'fontname':'Comic Sans Ms', 'fontsize':20})
plt.xlabel('Ages',fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})
plt.ylabel('Median Salary', fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})


plt.tight_layout()
plt.show()
```

```python
#using conditional filling   

plt.plot(ages,dev_salaries,color = '#444444',linestyle ='--',label = 'Alldevs')
plt.plot(ages,py_salaries,label = 'Python')

overall_median = 52787

#the third arguement speicifies where exactly we want our fill to cut through
#where -> where we want to fill specifying conditional 
plt.fill_between(ages,py_salaries, overall_median, 
                 where = (py_salaries > overall_median), 
                 interpolate  =True, alpha = 0.25)  

plt.fill_between(ages,py_salaries, overall_median, 
                 where = (py_salaries <= overall_median), 
                 interpolate  =True,color = 'red', alpha = 0.25)  

plt.legend()

plt.title('Median Slary(USD) by Ages', fontdict={'fontname':'Comic Sans Ms', 'fontsize':20})
plt.xlabel('Ages',fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})
plt.ylabel('Median Salary', fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})

plt.tight_layout()
plt.show()
```

```python
#filling the area between our line plot
# in this case -> betweeen our pythons salaries and our developer salaries


plt.plot(ages,dev_salaries,color = '#444444',linestyle ='--',label = 'Alldevs')
plt.plot(ages,py_salaries,label = 'Python')

overall_median = 52787

#the third arguement speicifies where exactly we want our fill to cut through
#where -> where we want to fill specifying conditional 
plt.fill_between(ages,py_salaries, dev_salaries, 
                 where = (py_salaries > dev_salaries), 
                 interpolate  =True, alpha = 0.25)  #fills between python line plot and all the way down to the xline 

plt.fill_between(ages,py_salaries, dev_salaries, 
                 where = (py_salaries <= dev_salaries), 
                 interpolate  =True, color = 'red', alpha = 0.25)  

plt.legend()

plt.title('Median Slary(USD) by Ages', fontdict={'fontname':'Comic Sans Ms', 'fontsize':20})
plt.xlabel('Ages',fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})
plt.ylabel('Median Salary', fontdict={'fontname':'Comic Sans Ms', 'fontsize':18})

plt.tight_layout()
plt.show()
```
